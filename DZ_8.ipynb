```python
#ЗАДАНИЕ 1

#Библиотека для запросов к сайтам через интернет
import requests

#Функция возвращает название валюты с максимальным значением курса
def valute_max():
    #Словарь для хранения информации курсе и названии валют
    value_and_name = {}

    #Перебираем информацию о каждой из валют и заисываем в словарь пару курс: название
    for valute in requests.get('https://www.cbr-xml-daily.ru/daily_json.js').json()['Valute'].values():
        value_and_name[valute['Value']] = valute['Name']

    #Сортируем словарь по ключам по убыванию и возвращаем значение от первого ключа
    return value_and_name[sorted(value_and_name, reverse=True)[0]]

#Проверка результатов
print(valute_max())
```

```python
#ЗАДАНИЕ 2

#Библиотека для запросов к сайтам через интернет
import requests

#Класс для работы с курсом валют
class Rate:
    #Инициализация переменных
    def __init__(self, format_ = 'value', diff = False):
        #Формат вывода информации о валюте: 'value' - курс, 'full' - полная информация
        self.format = format_
        #Выбор: True - выводить разницу курсов, False - последнее значение 
        self.diff = diff
    
    #Метод запрашивает информацию о валютах с сайта и возвращает полную информацию
    def exchange_rates(self):
        #Полная информация о валютах на текущую дату
        self.r_current = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return self.r_current.json()['Valute']
    
    #Метод возвращает информацию о валюте currency - полная информация о валюте при self.format = 'full' 
    #и только курс при self.format = 'value', а также разницу курсов при self.diff = True
    def make_format(self, currency):
        #Получаем полную информацию о валютах с помощью метода exchange_rates()
        response = self.exchange_rates()
        #Проверяем, есть ли валюта в общем списке с полной информацией о валютах
        if currency in response:
            #Если self.format = 'full', то возвращаем полную информацию по валюте
            if self.format == 'full':
                return response[currency]
            #Если self.format = 'value', то возвращаем курс валюты
            if self.format == 'value':
                #Если self.diff = True, то возвращаем разницу курсов
                if self.diff == True:
                    return response[currency]['Value'] - response[currency]['Previous']
                else:
                    return response[currency]['Value']
        return 'Error'
    
    def eur(self):
        """Возвращает курс евро на сегодня в формате self.format"""
        return self.make_format('EUR')
    
    def usd(self):
        """Возвращает курс доллара на сегодня в формате self.format"""
        return self.make_format('USD')
    
    def brl(self):
        """Возвращает курс бразильского реала на сегодня в формате self.format"""
        return self.make_format('BRL')

#Проверка результатов
r = Rate(format_='value', diff = True)
print(r.eur(), r.usd(), r.brl())
```

```python
#ЗАДАНИЕ 3

#Базовый класс для описания сотрудников
class Employee:
    #Инициализация переменных: имя сотрудника и пройденные аккредитации
    def __init__(self, name, seniority):
        self.name = name
        self.seniority = seniority
        #Начальный грейд сотрудника
        self.grade = 1
    
    def grade_up(self):
        """Повышает уровень сотрудника"""
        self.grade += 1
    
    def publish_grade(self):
        """Публикация результатов аккредитации сотрудников"""
        print(self.name, self.grade)
    
    def check_if_it_is_time_for_upgrade(self):
        pass

#Класс для дизайнеров    
class Designer(Employee):
    #Инициализация переменных: имя сотрудника и пройденные аккредитации (из родительского класса)
    def __init__(self, name, seniority, award = 2):
        #Вызываем метод __init__ родительского классса
        super().__init__(name, seniority)
        #Переменная количества наград
        self.award = award
        
    #Аккредитация сотрудника
    def check_if_it_is_time_for_upgrade(self):
        # для каждой аккредитации увеличиваем счетчик на 1
        # пока считаем, что все разработчики проходят аккредитацию
        self.seniority += 1
        # условие повышения сотрудника из презентации
        if self.seniority % 5 == 0:
            self.grade_up()
        # публикация результатов
        return self.publish_grade()
    
    #Получение премии
    def check_if_it_is_time_for_upgrade_award(self):
        #Добавляем премию
        self.award += 1
        #Условие повышения сотрудника: за каждые 7 баллов (1 премия = 2 балла)
        if (self.award * 2) % 7 <= 1 and self.award !=0:
            self.grade_up()
        # публикация результатов
        return self.publish_grade()
    
#Проверка результатов
alex = Designer('Александр', 0)
for i in range(20):
    alex.check_if_it_is_time_for_upgrade_award()
```
